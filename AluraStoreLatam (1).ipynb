### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import numpy as np
import matplotlib.pyplot as plt

#Ingreso por tienda
tiendas =[tienda, tienda2, tienda3, tienda4]
ingresos_tiendas = []
for i in range(len(tiendas)):
  ingresos_tiendas.append(tiendas[i]['Precio'].sum())
```

# 2. Ventas por categoría

```python
cols = tienda.columns
```

```python
#Armar el df donde se muestren las ventas organizadas para cada tienda por categoría
df_ventas_x_tienda = tienda.groupby('Categoría del Producto')[['Producto']].count().sort_values('Producto')
for i in range(len(tiendas)):
  strr='Ventas tienda'+ str(i+1)
  df_ventas_x_tienda[strr] = tiendas[i].groupby('Categoría del Producto')[['Producto']].count().sort_values('Producto')

df_ventas_x_tienda.drop('Producto', axis=1, inplace=True)
```

```python
df_ventas_x_tienda
```

# 3. Calificación promedio de la tienda


```python
df_calificación_x_tienda = tienda.groupby('Categoría del Producto')[['Calificación']].mean().sort_values('Calificación')
for i in range(len(tiendas)):
  strrr='Calificación tienda'+ str(i+1)
  df_calificación_x_tienda[strrr] = tiendas[i].groupby('Categoría del Producto')[['Calificación']].mean()

df_calificación_x_tienda.drop('Calificación', axis=1, inplace=True)
```

```python
df_calificación_x_tienda.describe()
```

# 4. Productos más y menos vendidos

```python

df_mas_vendidos_tienda = tienda.Producto.value_counts().to_frame().sort_values('count')
df_mas_vendidos_tienda2 = tienda2.Producto.value_counts().to_frame().sort_values('count')
df_mas_vendidos_tienda3 = tienda3.Producto.value_counts().to_frame().sort_values('count')
df_mas_vendidos_tienda4 = tienda4.Producto.value_counts().to_frame().sort_values('count')
```

# 5. Envío promedio por tienda

```python
#Armar el df donde se muestren los costos promedio de envio por categoria para cada tienda y de ahí podemos obtener el promedio para cada una
df_costosenvio_x_tienda = tienda.groupby('Categoría del Producto')[['Costo de envío']].mean()
for i in range(len(tiendas)):
  strr='CostosPromEnvio '+ str(i+1)
  df_costosenvio_x_tienda[strr] = tiendas[i].groupby('Categoría del Producto')[['Costo de envío']].mean()

df_costosenvio_x_tienda.drop('Costo de envío', axis=1, inplace=True)
```
